```python
# Through the heatmap, I found that Income, MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds, NumWebPurchases,
# NumCatalogPurchases columns have the most obvious relationship with NumStorePurchases. So then I wanna leveage regression plot
# as well as a simple regression sample to further explore.
```

```python
cols_2 = df4[['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
              'MntSweetProducts', 'MntGoldProds', 'NumWebPurchases', 'NumCatalogPurchases']].columns
len(cols_2)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12,12))
axes = axes.flatten()
for i, col in enumerate(cols_2):
    sns.regplot(df4, ax=axes[i], x=col, y='NumStorePurchases')
    axes[i].set_title(col)
for j in range(i+1, len(axes)):
    fig.delaxes[j]
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df5 = pd.concat([df4[cols_2], df4['NumStorePurchases']], axis=1)
df5.columns
results = {}
for col in df5.columns:
    if col != 'NumStorePurchases':
        x = df5[[col]]
        y = df5['NumStorePurchases']
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
        model = LinearRegression()
        model.fit(x_train, y_train)
        y_preds = model.predict(x_test)
        mse = mean_squared_error(y_test, y_preds)
        r2 = r2_score(y_test, y_preds)
        results[col] = {'Mean Squared Error':mse, 'R2 Score':r2}
results_df5 = pd.DataFrame(results).T
results_df5
```

```python
# From the results above, we can discover that MntWines has the highest R2 Score(0.41) and lowest Mean Squared Error(6.39), indicating that 
# MntWines has the strongest relationship with the number of store purchases amongst all, though it does not seem very significant. Variables like MntFruits,
# MntFishProducts, MntSweetProducts, MntGoldProds have high MSE and low R2 score, indicating they are weak preditors of NumStorePurchases. Generally
# speaking, the R2 score are all very low(<0.5), which means no single variable is strong preditor of NumStorePurchases.
```

```python
df4_filter = df4[df4['Year_Customer'].isin([2014, 2013])]
df4_filter['Year_Customer'].value_counts()
```

```python
Gold_mean = df4_filter['MntGoldProds'].mean()
Gold_mean
```

```python
Gold_above = df4_filter[df4_filter['MntGoldProds']>Gold_mean]
Gold_below = df4_filter[df4_filter['MntGoldProds']<=Gold_mean]
from scipy.stats import ttest_ind
Store_above = Gold_above['NumStorePurchases']
Store_below = Gold_below['NumStorePurchases']
t_stats, p_value = ttest_ind(Store_above, Store_below)
print('t-statistic: ', t_stats)
print('p-value: ', p_value)
```

```python
# The results indicate that there is robust evidence supporting a difference in NumStorePurchases between two MntGoldProds groups.
```

```python
Store_above_mean = Store_above.mean()
Store_below_mean = Store_below.mean()
print('Average store purchases(above average gold spending): ', Store_above_mean)
print('Average store purchases(below average gold spending): ', Store_below_mean)
```

```python
df4_filter['Gold_group'] = df4_filter['MntGoldProds'] > Gold_mean
plt.figure(figsize=(8,4))
sns.boxplot(df4_filter, x='Gold_group', y='NumStorePurchases', width=0.3)
plt.grid(axis='y', color='grey', linestyle='--', linewidth=0.5)
plt.xlabel('Gold spending group(below average: False, above average: True)', size=12)
plt.ylabel('Number of store purchases', size=12)
plt.title('The number of store purchases by gold spending group', size=14, weight='bold')
plt.show()
```

```python
# The boxplot support the hypothesis that individuals who spend more on gold tend to make more store purchases.
```

```python
Marital_PhD_Group = df2[df2['Marital_Education'] == 'Married PhD']
Other_Group = df2[df2['Marital_Education'] != 'Married PhD']
from scipy.stats import ttest_ind
Fish_Marital_PhD = Marital_PhD_Group['MntFishProducts']
Fish_Other = Other_Group['MntFishProducts']
t_stats, p_value = ttest_ind(Fish_Marital_PhD, Fish_Other)
print('t-statistic: ', t_stats)
print('p-value: ', p_value)
```
